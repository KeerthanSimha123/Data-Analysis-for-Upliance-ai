Number of cooking sessions per User ID and Meal Type

```python
a= x.groupby(['User ID', 'Meal Type_y'])['Session ID_x'].nunique().reset_index()
a
```

Renaming Session ID_x as Session_count

```python
a.rename(columns={'Session ID_x': 'Session Count'}, inplace=True)
a
```

Number of orders per User ID or Meal Type

```python
x=pd.merge(x,a, on=['User ID', 'Meal Type_y'], how='left')
x
```

```python
b=x.groupby(['User ID', 'Meal Type_x'])['Order ID'].nunique().reset_index()
```

```python
b.rename(columns={'Order ID': 'Order Count'}, inplace=True)
```

```python
x=pd.merge(x,b, on=['User ID', 'Meal Type_x'], how='left')
x
```

To find the relationship between cooking sessions and user orders, we need to find the correlation between them with value nearing to -1 indicates the -ve correlation and value nearing to 1 indicates the +ve correlation

correlation between Session Count and Order Count

```python
correlation = x[['Session Count', 'Order Count']].corr()
correlation
```

Correlation between session count and order amount

```python
correlation1=x[['Session Count', 'Amount (USD)']].corr()
correlation1
```

Demographic classification
1. Age


```python
labels = ['<30', '31-36', '37-40', '41+']
x['Age Group'] = pd.cut(x['Age'], bins=[0, 30, 36, 40, 100], labels=labels)

c= x.groupby('Age Group')['Dish Name'].value_counts().unstack()
c
```

```python
x.groupby('Meal Type_x')['Session Rating'].mean()
```

```python
v=x.groupby('Age Group')['Favorite Meal'].value_counts().unstack()
v
```

Exploratory Data Analysis aka Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x['Age Group'], x['Favorite Meal'])
plt.title('Favorite Meals by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
plt.pie(x['Favorite Meal'].value_counts(), labels=x['Favorite Meal'].value_counts().index, autopct='%1.1f%%')
plt.title('Favorite Meals')
plt.show()
```

```python
plt.pie(x.groupby('Meal Type_x')['Session Rating'].mean(), labels=x.groupby('Meal Type_x')['Session Rating'].mean().index,autopct='%1.1f%%')
plt.title('Average Session Rating by Meal Type')
plt.show()
```
